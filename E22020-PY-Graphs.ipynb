```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
def graph_v1(df):                                       #function takes one parameter--"filename.csv" file
    df = pd.read_csv(df)                                #this works only when file is in the working directory
    df_nm_features = df.select_dtypes(exclude='object')  #this variable will have all the numerical variable
    df_ct_features = df.select_dtypes(include='object')  #this one will have all the categorical

    
    for i in df_nm_features:                   #for loop for numerical varible to plot histogram & boxplot                                    
        plt.figure(figsize=(11,11))
        plt.subplot(2,1,1)                    #sublot takes three parameters row,column, index  
        sns.boxplot(x=df_nm_features[i],color='blue') #making boxplot of blue colour
        plt.xlabel(i,fontsize=11)
        plt.title('Boxplot of attribute '+i,fontsize=11) # string concatenation in naming
        plt.subplot(2,1,2)                    #this next graph will come below the above graph
        plt.hist(df_nm_features[i],color="blue",edgecolor="black") #making histogram of blue colour and black edge
        plt.tight_layout(pad=8)               #tight_layout is done to avoid graphs touch each in any manner 
        plt.xlabel(i)                         #labelling x axis
        plt.ylabel('Frequency ')
        plt.title('Histogram of attribute '+i,fontsize=11)
        plt.savefig(i+'.png')               #saving the graphs in working directory
        plt.show()
        
    for l in df_ct_features:
        df_ct_features[l].value_counts().plot(kind='bar',figsize=(5,5),color="blue",fontsize=11) #barplot for categoricalvar
        plt.xlabel(l)
        plt.title('Bar graph of attribute  '+l,fontsize=11)
        plt.savefig(l+'.png')               #saving the graphs in working directory
        plt.show()
```

```python
graph_v1("kc_house_train_data.csv")
```

```python
def graph_v2(df):                         #this func also takes one parameter--"filename.csv"
    df = pd.read_csv(df)                  #this works only when file isin the working directory
    df_nm_features = df.select_dtypes(exclude='object')
    df_ct_features = df.select_dtypes(include='object')
    
    for i in df_nm_features:
        if len(df_nm_features[i].unique())>=5:  #it checks whether an numerical attribute has 5 more than 5 unique variables
            plt.figure(figsize=(11,11))         #if the above condition is satisfied, it plots boxplots and histograms
            plt.subplot(2,1,1)
            sns.boxplot(x=df_nm_features[i],color='blue')
            plt.xlabel(i,fontsize=11)
            plt.title('Boxplot of attribute '+i,fontsize=11)
            plt.subplot(2,1,2)
            plt.hist(df_nm_features[i],                              
                     color="blue",
                     edgecolor="black")
            plt.tight_layout(pad=8)
            plt.xlabel(i)
            plt.ylabel('Frequency ')
            plt.title('Histogram of attribute '+i,fontsize=11)
            plt.savefig("v2"+i+'.png')
            
        else:          #above condition unsatisfied then it is treated as descrete numerical variable and bar graph is made
            df_nm_features[i].value_counts().plot(kind='bar',figsize=(5,5),
                                                       color="blue",fontsize=11)
            plt.xlabel(i)
            plt.title('Bar graph of attribute  '+i,fontsize=11)
            plt.savefig("v2"+i+'.png')
            
            

        
    for l in df_ct_features:
        df_ct_features[l].value_counts().plot(kind='bar',figsize=(5,5),
                                                       color="blue",fontsize=11)
        plt.xlabel(l)
        plt.title('Bar graph of attribute  '+l,fontsize=11)
        plt.savefig("v2"+l+'.png')
        
```

```python
graph_v2("kc_house_train_data.csv")
```

```python
def graph_v3(df,columns=None): #func takes two args one is filename and next is columns whose graphs to be plotted
    df = pd.read_csv(df)
    if columns==None:     #if default i.e none is given then all the columns or complete dataset is taken
        df=df.iloc[:]
    else:
        df=df[columns]     #otherwise the specified columns are dealt with
    
    df_nm_features = df.select_dtypes(exclude='object')
    df_ct_features = df.select_dtypes(include='object')
    
    for i in df_nm_features:
        if len(df_nm_features[i].unique())>=5:
            plt.figure(figsize=(11,11))
            plt.subplot(2,1,1)
            sns.boxplot(x=df_nm_features[i],color='blue')
            plt.xlabel(i,fontsize=11)
            plt.title('Boxplot of attribute '+i,fontsize=11)
            plt.subplot(2,1,2)
            plt.hist(df_nm_features[i],                              
                     color="blue",
                     edgecolor="black")
            plt.tight_layout(pad=8)
            plt.xlabel(i)
            plt.ylabel('Frequency ')
            plt.title('Histogram of attribute '+i,fontsize=11)
            plt.savefig('v3'+i+'.png')
            
        else:
            df_nm_features[i].value_counts().plot(kind='bar',figsize=(5,5),color="blue",fontsize=11)
            plt.xlabel(i)
            plt.title('Bar graph of attribute  '+i,fontsize=11)
            plt.savefig('v3'+i+'.png')
            
            

        
    for l in df_ct_features:
        df_ct_features[l].value_counts().plot(kind='bar',figsize=(5,5),color="blue",fontsize=11)
        plt.xlabel(l)
        plt.title('Bar graph of attribute  '+l,fontsize=11)
        plt.savefig('v3'+l+'.png')
       
```
